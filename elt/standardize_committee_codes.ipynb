```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = 100
from congress import Congress
```

Since session 113, Propublica assigned codes to each committee. This notebook will make a crosswalk from committee name: code for all sessions in dataset. Will attempt to match unstandardized names

```python
# get master list of committees/codes
propublica = Congress('apikey')
comm_master_list = []
for session in range(110, 116):
    for body in ['house','senate', 'joint']:
        uri = "{}/{}/committees.json".format(session, body)
        response = propublica.fetch(uri)
        for committee in response['committees']:
            row = {}
            row['code'] = committee['id']
            row['name'] = committee['name']
            row['body'] = body
            row['session'] = session
            comm_master_list.append(row)
```

```python
comm_master_list = pd.DataFrame(comm_master_list)
comm_master_list = comm_master_list.drop_duplicates(subset = ['body','code', 'name'])
comm_master_list
```

```python
subjects = []
for i, row in comm_master_list.iterrows():
    subject = row['name'].lower().replace("committee", "")
    subject = subject.replace(' on', "")
    subject = subject.replace('joint ', "")
    subject = subject.replace(",", "")
    subject = subject.replace("'", "")
    subject = subject.replace("the ", "")
    subject = subject.replace(" and", "")
    subject = subject.replace("  ", " ")
    subjects.append(subject)
```

```python
comm_master_list['subject'] = subjects
```
