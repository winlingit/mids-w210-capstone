```python
# get number of candidates each PAC contributed to in all election cycles
df_all = pd.read_csv('preprocessed/campaign_finance/pac_direct_all.csv', encoding='latin1')
subset = ['Cycle','PACID','CID']
df_all = df_all.drop_duplicates(subset=subset)[subset]
df_activity = df_all.groupby(['PACID','Cycle']).size().unstack()
df_activity_z = (df_activity - df_activity.mean())/df_activity.std()
df_activity_z.to_csv('preprocessed/campaign_finance/pac_cand_count_norm.csv')
df_activity.fillna(0, inplace=True)
df_activity.to_csv('preprocessed/campaign_finance/pac_cand_count.csv')
```

```python
def compare_contributions(candidates, 
                          df, 
                          pac_frequency,
                          industry = None):
    """Compares PAC contributions between two candidates 
        (In the context of this analysis, that is member and a bill sponsor)
        
        Args:
            candidates: list of two candidate ids (CID field)
                First should be the member, second is the sponsor
            df: the dataframe that contains the PAC contributions for those candidates
            pac_frequency: Series of number of normalized candidates PAC contributes to for an election year
                this is saved as 'preprocessed/campaign_finance/pac_cand_count_norm.csv'
            industry: Only look at similarities for PACs in that industry. Optional
            
        Returns:
            Dictionary of measures of similarity between the two candidates:
                member_cid: candidates[0]
                sponsor_cid: candidates[1]
                n_incommon: Number of contributing PACs in common
                n_incommon_weighted: Number of contributing PACs in common weighted by total PAC activity
                n_member_total: Total number of PACs for member
                n_sponsor_total: Total number of PACs for sponsor
                percent_sponsor: n_incommon / n_sponsor_total
                percent_member: n_incommon / n_member_total
                ammount_correlation: correlation of total contribution amount of PACs incommon
                contributions_member: total amount of contributions to member
                contributions_sponsor: total amount of contributions to sponsor
        """
    df = df[df.CID.isin(candidates)]
    contributers = []
    for candidate in candidates:
        contributers.append(set(df[df.CID == candidate].PACID))
    
    # separate df into venn diagram of contributers
    df_member = df[df.PACID.isin(list(contributers[0] - contributers[1]))]
    df_sponsor = df[df.PACID.isin(list(contributers[1] - contributers[0]))]
    df_incommon = df[df.PACID.isin(list(contributers[1] & contributers[0]))]
    
    output = {}
    output['member_cid'] = candidates[0]
    output['sponsor_cid'] = candidates[1]
    
    # append venn diagram counts into output dictionary
    output['n_incommon'] = len(df_incommon.PACID.unique())
    output['n_member_total'] = output['n_incommon'] + len(df_member.PACID.unique())
    output['n_sponsor_total'] = output['n_incommon'] + len(df_sponsor.PACID.unique())
    output['percent_sponsor'] = output['n_incommon'] / output['n_sponsor_total']
    output['percent_member'] = output['n_incommon'] / output['n_member_total']
    
    # pacs in common weighted by document frequency
    pac_frequency = pac_frequency.reset_index()
    pac_frequency.columns.values[1] = 'doc_freq'
    df_incommon = pd.merge(df_incommon, pac_frequency, how = 'left', on = 'PACID')
    try:
        output['n_incommon_weighted'] = df_incommon.groupby('PACID').doc_freq.mean().sum() / len(df_incommon.groupby('PACID'))
    except:
        pass
    
    
    # cash values
    try:
        output['amount_correlation'] = df_incommon.groupby(['PACID', 'CID']).Amount.sum().unstack().corr().iloc[1,0]
    except:
        output['amount_correlation'] = np.nan
    output['contributions_member'] = df_member.Amount.sum() + df.groupby(['CID']).Amount.sum()[candidates[0]]
    output['contributions_sponsor'] = df_sponsor.Amount.sum() + df.groupby(['CID']).Amount.sum()[candidates[1]]
    
    return output
#test = compare_contributions(['N00033508', 'N00035403'], df_16, df_activity_z[2016])
```

```python
path = 'preprocessed/campaign_finance/pac_direct_'
fname = 'preprocessed/campaign_finance/mem_relationship_'

for year in reversed(np.linspace(1990, 2016, 14)):
    print(year)
    output = []
    year = int(year)
    df = pd.read_csv(path + str(year) + '.csv', encoding='latin1')
    for member, sponsor in combinations(df.CID.unique(),2):
        row = compare_contributions([member, sponsor], df, df_activity_z[year])
        output.append(row)
    
    output1 = pd.DataFrame(output)
    rev = {'contributions_member':'contributions_sponsor', 
       'contributions_sponsor':'contributions_member', 
       'member_cid':'sponsor_cid',
       'n_member_total':'n_sponsor_total', 
       'n_sponsor_total':'n_member_total', 
       'percent_member':'percent_sponsor', 
       'percent_sponsor':'percent_member',
       'sponsor_cid':'member_cid'}
    output2 = output1.rename(columns=rev)
    output3 = pd.concat([output1, output2])
    output3.to_csv(fname + str(year) + '.csv', index = False)
```
